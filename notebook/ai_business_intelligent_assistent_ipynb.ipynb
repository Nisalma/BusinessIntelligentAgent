```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv', encoding='ISO-8859-1')
```

```python
print(f"Rows before cleaning: {len(df)}")
```

```python
df = df.dropna(subset=['CustomerID', 'Description'])
```

```python
print(f"Rows after dropping nulls: {len(df)}")
```

```python
#Fix types make sure customer id is a whole number not a decimal like 12345.0
df['CustomerID']=df['CustomerID'].astype(int)
```

```python
# 4. FIX DATES: Convert the text date into a real Python Date object
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
# 5. FEATURE ENGINEERING: Calculate Revenue per row
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']
```

```python
# 6. FILTER: Keep only positive sales (we can analyze returns later if we want)
df_clean = df[df['Quantity'] > 0].copy()
```

```python
# 7. SAVE: Export this as a new, perfect file
df_clean.to_csv('Clean_Online_Retail.csv', index=False)
```
